```python
%matplotlib widget
import seaborn as sns
import pandas as pd
import numpy as np


df = pd.read_csv('./out.csv', dtype=float);
#df['c'] = math.sqrt(df['x']**2 + df['y']**2)
#df['pdist'] = np.sqrt(df['x']**2 + df['y']**2)
df['xp'] = df['x'].shift(1).ffill()
df['xp'][0] = 0

df['yp'] = df['y'].shift(1).ffill()
df['yp'][0] = 0

df['hue'] = np.sqrt( (df['x'] - df['xp'])**2 + (df['y'] - df['yp'])**2 )
#df['dist'] = df['pdist'].shift(1).ffill()
#df['dist'][0] = 0
#df['hue'] = df['pdist'] - df['dist']
df
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(11, 11))
sns.scatterplot(data=df, x='x', y='y', hue='hue', alpha=0.1, size='x', sizes=(1, 1))
```
